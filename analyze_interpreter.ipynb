```python
import pandas as pd
from collections import defaultdict
import ast
```

```python
df = pd.read_csv('/archetype-predict-pkg/results/january24_word_attributions.csv')
df.head()
```

for each column
    for each row
        read list of tuples
            add the tuples to the dictionary
                if the key is already in the dictionary, average the values
                else add the key and value to the dictionary

```python


# Iterate through each column
for column in df.columns:
    # Dictionary to store cumulative sums and counts for averaging
    token_stats = defaultdict(lambda: {"sum": 0, "count": 0})
    # Iterate through each row in the column
    for row in df[column]:
        #change row type to list
        row = ast.literal_eval(row)
        if isinstance(row, list):
            for item in row:

                if isinstance(item, tuple) and len(item) == 2:
                    token, attribution = item
                    token_stats[token]["sum"] += attribution
                    token_stats[token]["count"] += 1

    #exclude tokens with less than 12 counts
    token_stats = {token: stats for token, stats in token_stats.items() if stats["count"] > 12}

# Calculate average attribution for each token
    average_attributions = {token: stats["sum"] / stats["count"] for token, stats in token_stats.items()}

    # Display results
    print("Average attributions for each token:")
    # Sort the tokens by average attribution and print first 10 results
    for token, avg in sorted(average_attributions.items(), key=lambda x: x[1], reverse=True)[:10]:
    #for token, avg in sorted(average_attributions.items(), key=lambda x: x[1], reverse=True):
        print(f"{token}: {avg}")    
```
